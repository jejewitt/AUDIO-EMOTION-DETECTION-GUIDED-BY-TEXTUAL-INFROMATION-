```python
import librosa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
import scipy.io.wavfile as wav
import scipy.signal as signal
import seaborn as sns
import librosa.display
```

```python
path  = './Data/Actor_01/03-01-01-01-01-01-01.wav'
y, sr = librosa.load(path, mono=True)
mfcc = librosa.feature.mfcc(y=y, sr=sr)


plt.title('MFCC, Neutral Emotion, Actor 01')

librosa.display.specshow(mfcc, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python

path  = './Data/Actor_01/03-01-01-01-01-01-01.wav'
y, sr = librosa.load(path, mono=True)

X = librosa.stft(y)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(20, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.title('Decibel Scaled Spectrogram Neutral Emotion Actor 01 ')

plt.colorbar()
```

```python
wav_list = []
labels = []
for i in range(1,25):
    da= os.listdir(f"Data/Actor_%02d"% (i,))
    # print(da[0])
    
    for k in da:
        if '.wav' in k:
            labels.append(k[6:8])
            wav_list.append(f"./Data/Actor_%02d/"% (i,) + k) 
```

```python
import matplotlib.pyplot as plt

x = ['neutral','calm','happy','sad','angry','fearful','disgust','surprised']
y = [96,192,192,192,192,192,192,192]
fig, ax = plt.subplots()
one = ax.bar(x, y, label='Tweets labeled per Emotion')

plt.xlabel('Emotion')
plt.ylabel('Samples')
ax.set_title('Number of Samples per Emotion')
# ax.bar_label(one ,padding=3)

plt.tight_layout(pad=0.5)
plt.margins(0.1)
fig.autofmt_xdate() # make space for and rotate the x-axis tick labels

plt.show()



```

```python

dur = np.zeros(len(wav_list))
for i,val in enumerate(wav_list):
    dur[i]=librosa.get_duration(filename=val)

```

```python
print('mean: ',dur.mean())
print('standard Deviation: ',dur.std())

```
